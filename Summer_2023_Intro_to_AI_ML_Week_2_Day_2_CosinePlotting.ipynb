```python
#import numpy & plotting tools
import numpy as np
import matplotlib.pyplot as plt

# Make time variable
t = np.arange(0,2,0.01)

#print(t)

#specify frequency of cosine
f = 1

# Make cosine
y = np.cos(2*np.pi*f*t)
plt.plot(t,y, label = 'Original')

# Print shape of t
print(t.shape[0])

# Add noise to y
z = np.random.normal(0,0.1,t.shape[0])
#plt.plot(t,z)
#plt
y_noisy = y+z
plt.plot(t,y_noisy, label = 'Noisy')

# Import scipy
from scipy import signal

# Make filter, apply filter, and run
b, a = signal.butter(5,0.1,'low')
y_filtered = signal.filtfilt(b,a,y_noisy)
plt.plot(t, y_filtered, label = 'Filtered')
plt.legend()
```
